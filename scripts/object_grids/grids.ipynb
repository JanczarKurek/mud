```python
from PIL import Image, ImageDraw
```

```python
class SpriteGrid:
    def __init__(self, tilesize: int, xcnt: int, ycnt: int):
        self._tilesize = tilesize
        self._xcnt = xcnt
        self._ycnt = ycnt
        self._sizex = xcnt * tilesize
        self._sizey = ycnt * tilesize
    def tile_rec(self, x, y, offset=0):
        return [
            (self._tilesize * x + offset, self._tilesize * y + offset),
            (self._tilesize * (x + 1) - 1 + offset, self._tilesize * (y + 1) - 1 + offset)
        ]
    def draw(self):
        colors1 = ("red", "green")
        colors2 = ("blue", "magenta")
        brush_size = 2
        def style(c, x, y):
            return {
                "outline": c[(x + y)%2],
                "width": brush_size,
            }
        img = Image.new("RGBA", (self._sizex, self._sizey))
        draw = ImageDraw.Draw(img)
        for x in range(self._xcnt):
            for y in range(self._ycnt):
                draw.rectangle(self.tile_rec(x, y), **style(colors1, x, y))
        for x in range(self._xcnt - 1):
            for y in range(self._ycnt - 1):
                draw.rectangle(self.tile_rec(x, y, offset = self._tilesize//2), **style(colors2, x, y))
        return img
```

```python
grid = SpriteGrid(48, 4, 6)
grid.draw().show()
```
